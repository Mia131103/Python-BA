### Finding data

function = locate 
we can write any kind of conditional statement with it

Works in both rows and columns

syntax: tablename.loc[rows(row number/range/row conditions), columnname]

Python uses & for and

Python uses | for or

```python
product_table.loc[5] # shows all the record in the 5th row
```

```python
product_table.loc[5:10]  #shows rows from 5 to 10
```

```python
product_table.loc[5:10, ['ProductKey','ProductName','Color']]
#We can specify the columns as well
```

```python
product_table.loc[[5,6,10,55], ['ProductKey','ProductName','Color']] 
#We can specify both rows and columns
```

```python
#If we only want to see the products whos color is black
product_table.loc[(product_table['Color']=='Black'), ['ProductKey','ProductName','Color']]
```

```python
product_table.loc[(product_table['StandardCost'] <= 50), ['ProductKey','ProductName']]
```

Q. Find those records in the data where the listprice is less than 40 and columns name should be List price, size, size range, weight

```python
product_table.loc[(product_table['ListPrice'] < 40), ['ListPrice','Size','SizeRange','Weight']]
```

```python
#row 397
product_table.loc[(product_table['ProductKey'] == 397), ['ProductKey','ProductName','Color']]
```

Q. Extract all rows where product is not black and size is greater than 40 and weight is less than 50. Dsiplay all columns.

```python
product_table.loc[(product_table['Color'] != 'Black') & (product_table['Size'] > 40) & (product_table['Weight'] < 50)]
```

```python
product_table.dropna() #deletes all rows and columns with missing values in the table
```

```python
product_table.drop_duplicates()  #deletes all recurring values
```

```python
product_table.drop('SafetyStockLevel', axis = 1, inplace = True) 

#columns means axis = 1, rows means axis = 0
#shows error bcz it has been run once and now there is no such column
```

```python
product_table.drop(2, axis = 0, inplace = True) 
#deleted row 2
```
