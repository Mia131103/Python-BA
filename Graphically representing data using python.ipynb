### Graphs 


```python
import matplotlib.pyplot as plt
```

```python
a = [1,2,3,4,5]
b = [2,4,6,8,10]
```

```python
plt.plot(a, b)  # a corresponds to x axis and b corresponds to y axis
```

```python
plt.scatter(a, b)
```

```python
a = [1,2,3,4]
b = [2,4,6,8,10]
plt.scatter(a, b)
#We need to have the same number of values for both a and b
```

```python
a = [1,2,3,4,5]
b = [2,4,6,8,10]
plt.scatter(a, b)
plt.plot(a, b)
```

```python
c = [4,7,11,9,5]
d = [65,23,78,53,13]
plt.scatter(a, b)
plt.scatter(c, d)
plt.plot(d, a)   
#We can have all the graphs on the same figure to make it easy to compare
plt.xlabel('Sales')
plt.ylabel('Volume')
plt.title("A comparison of sales and volume")
```

```python
plt.scatter(a, b, linewidths = 7, color = 'red', marker = 'x')
#can customize it anyway we want
plt.savefig('banana', dpi = 500) #to downnload graph to computer. dpi represents quality of graph
plt.annotate('very important', (2.5,8)) #can write text anywhere on the graph by giving the coordinate of the place we want
```

```python
plt.pie(a, labels = [1,2,3,'problem',5], explode = (0,0,0,0.1,0)) 
plt.show() #just like slideshow, we will only see graph, without its metadata
#explode means to specify which section has to be seperated and by how much
```

```python
plt.boxplot([a,b,c])
plt.show()
```
