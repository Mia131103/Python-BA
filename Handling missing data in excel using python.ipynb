### Missing data handling
* delete row
* replace row with mean/mode
* interpolatae
* extropolate

For accessing each collums, write 'tablename'['columname'].function()

```python
import pandas as pd
product_table = pd.read_excel(r'Adventure Works 2020.xlsx', sheet_name = 'Products')
product_table
```

```python
product_table.isnull().sum()         #shows the sum of all the missing values in the table
```

Replacing missing values with a particular number:

inplace = true permanentally changes your data in the notebook, not in the actual sheet; python remembers the change in the rest of the code

```python
product_table['StandardCost'].fillna(10, inplace=True) #fills 10 in place of missing values in the standard cost column 
```

Replacing missing values with the average

```python
listpricemean = product_table['ListPrice'].mean() 
listpricemean
```

```python
#Now to replace the missing values with this average:
product_table['ListPrice'].fillna(listpricemean, inplace = True)
```

Replacing missing values with mode

```python
Color_mode = product_table['Color'].mode()
print(Color_mode)
product_table['Color'].fillna(str(Color_mode), inplace = True)
#we can't write Color_mode directly in place of black as the datatype of the output of Color_mode is object instead of string
```
